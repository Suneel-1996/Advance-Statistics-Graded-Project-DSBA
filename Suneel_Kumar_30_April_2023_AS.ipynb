#Problem 3:

The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain; Answer the questions below based on the given information; (Provide an appropriate visual representation of your answers, without which marks will be deducted)





##3.1 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?

###Method 1


```python
x=np.linspace(-5,5,1000)

y=stats.norm.pdf(x,0,1)
mu=5
sigma=1.5
plt.plot(x,y)
critical_value=(3.17-mu)/sigma
print("The proportion of gunny bags with breaking strength less than 3.17 kg per sq cm is : ",round((stats.norm.cdf(critical_value))*100,2),"%")
plt.axvline(critical_value,c='g')
x1=np.linspace(-5,critical_value,100)
plt.fill_between(x1,stats.norm.pdf(x1),color='r')

plt.show()
```

###Method 2

```python
mu=5
sigma=1.5
print("The proportion of gunny bags with breaking strength less than 3.17 kg per sq cm is : ",round(stats.norm.cdf(3.17,mu,sigma)*100,2),"%")

x=np.random.normal(loc=5,scale=1.5,size=1000)
sb.histplot(x,kde=True)
plt.axvline(3.17,color='r')
plt.fill_betweenx(x,3.17,color='r')
plt.show()
```

#Problem 4:

 

Grades of the final examination in a training course are found to be normally distributed, with a mean of 77 and a standard deviation of 8.5. Based on the given information answer the questions below.

 



##4.1 What is the probability that a randomly chosen student gets a grade below 85 on this exam?



```python
#Given statement has information related to population mean(mu) and standard deviation(sigma).
mu=77
sigma=8.5
zscore_85=(85-mu)/sigma
#Method1
print("Probability that a randomly chosen student gets grade below 85 is : ",round(stats.norm.cdf(zscore_85)*100,2),"%")
#Method2
print(stats.norm.cdf(85,loc=mu,scale=sigma))
x=np.linspace(-8.5,8.5,100)
y=stats.norm.pdf(x,0,1)
plt.plot(x,y)
plt.axvline(zscore_85,c='g')
x1=np.linspace(-8.5,zscore_85,100)
plt.fill_between(x1,stats.norm.pdf(x1),color='g',hatch='-')
plt.show()

```

##4.2 What is the probability that a randomly selected student scores between 65 and 87?



```python
zscore_65=(65-mu)/sigma
zscore_87=(87-mu)/sigma
#Method1
print("Probability that a randomly chosen student score between 65 and 87 is : ",round((stats.norm.cdf(zscore_87)-stats.norm.cdf(zscore_65))*100,2),"%")
#Method2
print(stats.norm.cdf(87,loc=mu,scale=sigma)-stats.norm.cdf(65,loc=mu,scale=sigma))
x=np.linspace(-8.5,8.5,100)
y=stats.norm.pdf(x,0,1)
plt.plot(x,y)
plt.axvline(zscore_65,c='g')
plt.axvline(zscore_87,c='g')
x1=np.linspace(zscore_65,zscore_87,100)

plt.fill_between(x1,stats.norm.pdf(x1),color='g')

plt.show()

```

##4.3 What should be the passing cut-off so that 75% of the students clear the exam?

```python
#75 % students means the distqrubution above 25 %, below is the solution
print("Passing cut-off so that 75 % of the students clear the exam is : ",stats.norm.ppf(.25,mu,sigma))
x=np.linspace(-8.5,8.5,100)
y=stats.norm.pdf(x,0,1)
plt.plot(x,y)
cvalue=stats.norm.ppf(.25)
plt.axvline(cvalue,c='r')

plt.show()
```
